### Get the the BQ data to create a fake analytical file just for small markets

```python
# get daily data 
daily_df = GetBQData(config_dict["PROPHETSETTING"]["training_query"], config_dict["PROPHETSETTING"]["input"], config_dict["PROPHETSETTING"]["output"])


df = daily_df[daily_df['market'].str.contains(small_market_str)].copy() # filter data with only the markets inside the SM group
df = df.groupby(['region', 'market', 'account_type', 'funnel', 'ds']).sum().reset_index() # group by to remove any duplicate values
df = df.rename(columns={'ds':'date', 'X':'cost', 'y':'conversion_quantity'}) # rename columns 
df.date = pd.to_datetime(df.date) #ensure date column is a date type 

date_index = pd.date_range(df.date.min(), df.date.max()) # set a date index, all dates between min date and max date, to include all dates
multiple_index = pd.Index(date_index) # create the index

df = df.set_index(['date']) # set date as index

df = df.reset_index() # reset index 
df = df.fillna(1) # fill Nan values with 1
df.shape # get the shape of the data 
```

## Generate SM funnel splits

```python
# Get regional (SM) funnel split
regional_total_df = df[(df['funnel'].str.contains('Brand|Generic|Competitor'))].copy() # filter data to only the ones that have funnel as Brand, Generic, Competitor
regional_total_df = regional_total_df.groupby(['date']).sum(numeric_only=True).reset_index() # group by date 

regional_total_df.rename(columns = {'cost': 'total_cost', 'revenue': 'total_revenue', 'conversion_quantity': 'total_conversions'}, inplace = True) # rename columns
regional_total_df.tail() #print last obs 
```

Generate the percentage split for each funnel: Brand, Generic, Competitor. And get the spend and CQ for each funnel split (total_cost*percentage) 

```python
# set a filtered df with only brand funnel
Brand_df = df[(df['funnel'] == 'Brand')][['date', 'cost', 'conversion_quantity']].copy()
Brand_df = Brand_df.groupby(['date']).sum().reset_index()
Brand_df.rename(columns = {'cost': 'brand_cost', 'conversion_quantity': 'brand_conversion_quantity'}, inplace = True)

# set a filtered df with only Generic funnel
Generic_df = df[df['funnel'] == 'Generic'][['date', 'cost', 'conversion_quantity']].copy()
Generic_df = Generic_df.groupby(['date']).sum().reset_index()
Generic_df.rename(columns = {'cost': 'generic_cost', 'conversion_quantity': 'generic_conversion_quantity'}, inplace = True)

# set a filtered df with only Competitor funnel
Competitor_df = df[df['funnel'] == 'Competitor'][['date', 'cost', 'conversion_quantity']].copy()
Competitor_df = Competitor_df.groupby(['date']).sum().reset_index()
Competitor_df.rename(columns = {'cost': 'competitor_cost', 'conversion_quantity': 'competitor_conversion_quantity'}, inplace = True)

# merge the total regional df with each funnel split 
regional_split = regional_total_df.merge(Brand_df, how = 'left').merge(Generic_df, how = 'left').merge(Competitor_df, how = 'left')

# create the percentage funnel split for cost and CQ for each funnel
# create cost split percentage 
regional_split['brand_cost_split'] = regional_split['brand_cost']/regional_split['total_cost']
regional_split['generic_cost_split'] = regional_split['generic_cost']/regional_split['total_cost']
regional_split['competitor_cost_split'] = regional_split['competitor_cost']/regional_split['total_cost']
# create the CQ split percentage
regional_split['brand_conv_split'] = regional_split['brand_conversion_quantity']/regional_split['total_conversions']
regional_split['generic_conv_split'] = regional_split['generic_conversion_quantity']/regional_split['total_conversions']
regional_split['competitor_conv_split'] = regional_split['competitor_conversion_quantity']/regional_split['total_conversions']

regional_split = regional_split.fillna(0.33) # fill Nan values with 0.33, each split has the same weigth for those moments 
regional_split = regional_split[['date', 'brand_cost_split', 'brand_conv_split', 'generic_cost_split', 'generic_conv_split', 'competitor_cost_split', 'competitor_conv_split']] # subset to the data frame with just the percentage columns
regional_split
```

set the fake analytical file for the funnel splits. filter data between base_start and base_end, and add base_periods to create a base percentage split for the forecasting period

```python
regional_split.date = pd.to_datetime(regional_split.date) # ensure the date column is on date time format
temp1 = regional_split[(regional_split['date'] >= base_start) & (regional_split['date'] <= base_end)].copy() # filter data bewteen base_start and base_end periods
temp1['date'] = temp1['date'] + datetime.timedelta( days= base_periods) # add base_periods to current date to create the base percentage for each split on the forecast

regional_split_upd = pd.concat([temp1]) # create the regional (SM) split with the forecasted periods
regional_split_upd.rename(columns = {'date': 'ds'}, inplace = True) # rename date column to ds for an easier merge with forecasted df

idx = pd.date_range(regional_split_upd.ds.min(), regional_split_upd.ds.max()) # create the date range index
regional_split_upd = regional_split_upd.set_index('ds') # set date index
regional_split_upd = regional_split_upd.reindex(idx).fillna(0.333).rename_axis('ds').reset_index() # reindex and rename axis. fill Nan values with 0.33, if Nan each split weight is 0.33 
regional_split_upd
```

verify that for each date the sum of split percentage is 1

```python
regional_split_upd[['brand_cost_split', 'generic_cost_split', 'competitor_cost_split']].sum(axis=1).unique() # verify that 
```

### merge the SM funnel split percentage with the SM forecast and generate the forecast spend and CQ for each split

```python
# filter forecast to create a df with just SM 
sm_df = forecast_n_base_df[forecast_n_base_df['market'] == 'SM'].copy() # filter forecast with just SM markets 
sm_df = sm_df.merge(regional_split_upd, how = 'left', left_on = 'ds', right_on='ds') # merge the SM funnel split (regional_split) with the SM forecast (sm_df) 

# generate the SM brand split, multiplying the brand cost and CQ split percentage with the forecasted spend and CQ 
brand_df = sm_df.copy() 
brand_df['funnel'] = 'Brand'
brand_df['OptimalSpendValue'] = brand_df['OptimalSpendValue'] * brand_df['brand_cost_split']
brand_df['OptimalConversionQuantity'] = brand_df['OptimalConversionQuantity'] * brand_df['brand_conv_split']
brand_df['BaseSpend'] = brand_df['BaseSpend'] * brand_df['brand_cost_split']
brand_df['BaseConversionQuantity'] = brand_df['BaseConversionQuantity'] * brand_df['brand_conv_split']

# generate the SM generic split, multiplying the brand cost and CQ split percentage with the forecasted spend and CQ 
generic_df = sm_df.copy()
generic_df['funnel'] = 'Generic'
generic_df['OptimalSpendValue'] = generic_df['OptimalSpendValue'] * generic_df['generic_cost_split']
generic_df['OptimalConversionQuantity'] = generic_df['OptimalConversionQuantity'] * generic_df['generic_conv_split']
generic_df['BaseSpend'] = generic_df['BaseSpend'] * generic_df['generic_cost_split']
generic_df['BaseConversionQuantity'] = generic_df['BaseConversionQuantity'] * generic_df['generic_conv_split']

# generate the SM competitor split, multiplying the brand cost and CQ split percentage with the forecasted spend and CQ 
competitor_df = sm_df.copy()
competitor_df['funnel'] = 'Competitor'
competitor_df['OptimalSpendValue'] = competitor_df['OptimalSpendValue'] * competitor_df['competitor_cost_split']
competitor_df['OptimalConversionQuantity'] = competitor_df['OptimalConversionQuantity'] * competitor_df['competitor_conv_split']
competitor_df['BaseSpend'] = competitor_df['BaseSpend'] * competitor_df['competitor_cost_split']
competitor_df['BaseConversionQuantity'] = competitor_df['BaseConversionQuantity'] * competitor_df['competitor_conv_split']

# append each funnel split forecast to create a final SM dataframe forecast with a forecast spend and CQ for each funnel split 
sm_funnel_df = pd.concat([brand_df, generic_df, competitor_df])
sm_funnel_df.sum(numeric_only=1)
```

# Break SM results by market 

```python
# get small country split based on reference date range

small_country_split_df = df[df['market'].str.contains(small_market_str)&(df.date>=base_start)&(df.date<=base_end)].copy()

small_country_split_df.date = pd.to_datetime(small_country_split_df.date) # ensure the date column is a datetime type
small_country_split_df.date = small_country_split_df.date+datetime.timedelta(days=base_periods) #add base_perios to date to make it as an analytical file for the future periods
small_country_split_df = small_country_split_df.groupby(['market']).sum(numeric_only=1).reset_index() # group by market 

small_country_split_df['cost_split'] = small_country_split_df['cost']/small_country_split_df['cost'].sum(numeric_only=1) # get the cost split, market percentage over SM
small_country_split_df['revenue_split'] = small_country_split_df['revenue']/small_country_split_df['revenue'].sum(numeric_only=1) # get the renvenue split, market percentage over SM
small_country_split_df['conversion_split'] = small_country_split_df['conversion_quantity']/small_country_split_df['conversion_quantity'].sum(numeric_only=1) # get the conversion split, market percentage over SM
small_country_split_df.fillna(0, inplace = True)
small_country_split_df
```

```python
sm_final_df = pd.DataFrame()
for i in small_market_str.split('|'):
    sm_funnel_df_temp = sm_funnel_df.copy()
    sm_funnel_df_temp['market'] = i
    sm_funnel_df_temp['OptimalSpendValue'] = sm_funnel_df_temp['OptimalSpendValue'] * small_country_split_df.loc[small_country_split_df['market'] == i, 'cost_split'].array[0]
    sm_funnel_df_temp['OptimalConversionQuantity'] = sm_funnel_df_temp['OptimalConversionQuantity'] * small_country_split_df.loc[small_country_split_df['market'] == i, 'conversion_split'].array[0]
    sm_funnel_df_temp['BaseSpend'] = sm_funnel_df_temp['BaseSpend'] * small_country_split_df.loc[small_country_split_df['market'] == i, 'cost_split'].array[0]
    sm_funnel_df_temp['BaseConversionQuantity'] = sm_funnel_df_temp['BaseConversionQuantity'] * small_country_split_df.loc[small_country_split_df['market'] == i, 'conversion_split'].array[0]
    sm_final_df = pd.concat([sm_final_df, sm_funnel_df_temp])

sm_final_df = sm_final_df.drop(['generic_cost_split', 'competitor_cost_split', 'brand_cost_split', 'brand_conv_split', 'generic_conv_split', 'competitor_conv_split'], axis = 1)
sm_final_df = sm_final_df.rename(columns=lambda x: x[0:1].upper() + x[1:])



```

```python
LM_SM_three_funnel_df['Ds'] = pd.to_datetime(LM_SM_three_funnel_df['Ds'])
LM_SM_three_funnel_df['year_month'] = LM_SM_three_funnel_df['Ds'].dt.to_period('M')
```

```python
forecast_by_month = LM_SM_three_funnel_df.groupby('year_month').sum(numeric_only=1)
forecast_by_month.reset_index(inplace=True)
forecast_by_month['year_month'] = forecast_by_month['year_month'].dt.strftime('%Y-%m')

forecast_by_split = LM_SM_three_funnel_df.groupby(['year_month','Market','Account_type','Funnel']).sum(numeric_only=1)
forecast_by_split.reset_index(inplace=True)
forecast_by_split['year_month'] = forecast_by_split['year_month'].dt.strftime('%Y-%m')
```

```python
# Plot the bar graph
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot OptimalConversionQuantity
ax1.bar(forecast_by_month['year_month'], forecast_by_month['OptimalSpendValue'], color='blue', width=0.4, label='OptimalSpendValue', align='center')
# Plot BaseConversionQuantity
ax1.bar(forecast_by_month['year_month'], forecast_by_month['BaseSpend'], color='g', width=0.4, label='BaseSpend', align='edge')
ax1.set_ylabel('Spend')
ax1.set_xlabel('Year Month')
ax1.set_title('Optimal vs Base')
ax1.legend(loc='upper left')


ax2 = ax1.twinx()
ax2.plot(forecast_by_month['year_month'], forecast_by_month['OptimalConversionQuantity'], color='orange', label='OptimalConversionQuantity')
ax2.plot(forecast_by_month['year_month'], forecast_by_month['BaseConversionQuantity'], color='deepskyblue', label='BaseConversionQuantity')
ax2.set_ylabel('Conversion Quantity')
ax2.legend(loc='upper center')
ax2.tick_params(axis='y')

fig.tight_layout()

# Show the plot
plt.show()
```
